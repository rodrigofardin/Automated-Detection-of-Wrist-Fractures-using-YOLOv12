## Criado a Classe `YOLOWeightedDataset`

`YOLOWeightedDataset` irÃ¡ servir balancear o conjunto de dados durante o *data augmentation*

```python
from ultralytics.data.dataset import YOLODataset
import ultralytics.data.build as build

class YOLOWeightedDataset(YOLODataset):
    def __init__(self, *args, mode="train", **kwargs):
        """
        Dataset com balanceamento de classes via amostragem ponderada.
        """
        super(YOLOWeightedDataset, self).__init__(*args, **kwargs)

        self.train_mode = "train" in self.prefix

        # Contar instÃ¢ncias e calcular pesos
        self.count_instances()
        class_weights = np.sum(self.counts) / self.counts
        self.agg_func = np.mean

        self.class_weights = np.array(class_weights)
        self.weights = self.calculate_weights()
        self.probabilities = self.calculate_probabilities()

        # Exibir logs informativos
        self.show_class_distribution()

    def count_instances(self):
        """Conta quantas instÃ¢ncias existem por classe."""
        self.counts = [0 for _ in range(len(self.data["names"]))]
        for label in self.labels:
            cls = label['cls'].reshape(-1).astype(int)
            for id in cls:
                self.counts[id] += 1
        self.counts = np.array(self.counts)
        self.counts = np.where(self.counts == 0, 1, self.counts)

    def calculate_weights(self):
        """Calcula peso agregado para cada imagem baseado nas classes que ela contÃ©m."""
        weights = []
        for label in self.labels:
            cls = label['cls'].reshape(-1).astype(int)
            if cls.size == 0:
                weights.append(1)
                continue
            weight = self.agg_func(self.class_weights[cls])
            weights.append(weight)
        return weights

    def calculate_probabilities(self):
        """Converte os pesos para probabilidades normalizadas."""
        total_weight = sum(self.weights)
        probabilities = [w / total_weight for w in self.weights]
        return probabilities

    def show_class_distribution(self):
        """Exibe contagem, peso e probabilidade por classe."""
        class_names = self.data["names"]
        print("\nğŸ“Š DistribuiÃ§Ã£o de classes no dataset:")
        for i, name in enumerate(class_names):
            print(f"  Classe {i} ({name}): {self.counts[i]} instÃ¢ncias, peso={self.class_weights[i]:.4f}")

        print("\nğŸ¯ Probabilidades de amostragem por imagem:")
        print(f"  Min: {min(self.probabilities):.6f} | Max: {max(self.probabilities):.6f}")
        print(f"  MÃ©dia: {np.mean(self.probabilities):.6f}\n")

    def __getitem__(self, index):
        """Retorna imagem e label, aplicando amostragem ponderada no treino."""
        if not self.train_mode:
            return self.transforms(self.get_image_and_label(index))
        else:
            index = np.random.choice(len(self.labels), p=self.probabilities)
            return self.transforms(self.get_image_and_label(index))
```

```python
# Monkey patch method
build.YOLODataset = YOLOWeightedDataset
```

## Carregando o Modelo

```python
# Intanciando o modelo YOLOv12 L de um modelo prÃ©-treinado
#model = YOLO('yolov12l.pt')
model = YOLO("/kaggle/input/last-adamw-ep87/last.pt")

# Display model information (optional)
model.info()
```

## YOLO Fine-Tuning

```python
project_dir = 'yolov12l_adamw'
```

```python
# Treinamento do modelo no dataset GRAZPEDWRI-DX
results = model.train(
    # Caminho para o arquivo de configuraÃ§Ã£o do dataset
    data=new_yaml_path,

    # NÃºmero total de Ã©pocas de treinamento
    epochs=100,  # Aumentado para melhorar a convergÃªncia em fine-tuning

    # Tamanho do batch (ajustado para lidar limitaÃ§Ã£o de memÃ³ria)
    batch=16,

    # Tamanho da imagem de entrada (em pixels)
    imgsz=640,  # MantÃ©m o padrÃ£o usado no prÃ©-treinamento

    # GPU a ser usada (0 = primeira GPU)
    device=[0,1],

    # === Otimizador e taxa de aprendizado ===
    optimizer='AdamW',
    lr0=1e-3,
    momentum=0.937,
    weight_decay=5e-4,
    patience=50,
    amp=True,
    resume=True,

    name=project_dir, # Nome do experimento (cria uma pasta com esse nome para salvar logs/pesos)
)
```
